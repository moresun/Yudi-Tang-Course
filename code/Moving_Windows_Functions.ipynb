```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.Series(np.random.randn(600),index=pd.date_range('7/1/2016',freq='D',periods=600))
```

```python
df.head()
```

```python
r=df.rolling(window=10)
r
```

```python
print(r.mean())
```

```python
plt.figure(figsize=(15,5))
df.plot(style='r--')
df.rolling(window=10).mean().plot(style='b')
```
